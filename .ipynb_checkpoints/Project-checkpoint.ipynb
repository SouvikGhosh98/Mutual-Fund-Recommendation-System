## One Hot Encoding
Converting sub_category into 1s ans 0s to do clustering.

```python
reduced_data_encoded = pd.get_dummies(reduced_data, columns=['sub_category'], prefix=['sub_category'])
reduced_data_encoded.head()
```

```python
reduced_data_encoded = reduced_data_encoded*1
reduced_data_encoded.head()
reduced_data_encoded.info()
```

```python
columns_to_convert = ['sortino', 'alpha', 'sd', 'beta', 'sharpe']

for column in columns_to_convert:
    reduced_data_encoded[column] = pd.to_numeric(reduced_data_encoded[column], errors='coerce')

# Print the converted DataFrame
print(reduced_data_encoded)
reduced_data_encoded.info()
```

## Removing NaN Values

```python
reduced_data_encoded.dropna(inplace=True)
reduced_data_encoded.info()
reduced_data_encoded.head()
```

## Elbow Method for 1yr Returns

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.scatter(returns_1yr['risk_level'], returns_1yr['returns_1yr'], c='blue', alpha=0.5)
plt.title('Risk Level vs 1-Year Returns')
plt.xlabel('Risk Level')
plt.ylabel('1-Year Returns')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

wcss = []

K = range(1, 30)
for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(returns_1yr)
    wcss.append(kmeans.inertia_)  

# Plot the Elbow Method graph
plt.figure(figsize=(10, 6))
plt.plot(K, wcss, 'bx-')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Within-Cluster Sum of Squares (WCSS)')
plt.title('Elbow Method For Optimal k')
plt.xticks(np.arange(min(K), max(K)+1, 1))
plt.grid(True)
plt.show()
```

## Corelation Matrix

```python
risk_returns_all_yrs = reduced_data_encoded[['risk_level', 'returns_1yr', 'returns_3yr', 'returns_5yr']]
risk_returns_all_yrs.info()
```

```python
# from sklearn.metrics.pairwise import cosine_similarity

# cosine_sim_matrix = cosine_similarity(risk_returns_all_yrs)
# cosine_sim_data = pd.DataFrame(cosine_sim_matrix, index=risk_returns_all_yrs.index, columns=risk_returns_all_yrs.index)
# print(cosine_sim_data)

from sklearn.metrics.pairwise import cosine_similarity
from sklearn.preprocessing import StandardScaler

risk_returns_all_yrs_array = risk_returns_all_yrs.to_numpy()

# Separate risk and returns columns
risk = risk_returns_all_yrs_array[:, 0]
returns = risk_returns_all_yrs_array[:, 1:]

# Normalize the returns data
scaler = StandardScaler()
returns_normalized = scaler.fit_transform(returns)

# Combine normalized returns with risk
data_normalized = np.column_stack((risk, returns_normalized))

# Compute cosine similarity matrix
cosine_sim_matrix = cosine_similarity(data_normalized)

print("Cosine Similarity Matrix:")
print(cosine_sim_matrix)
```

```python
wcss = []

K = range(1, 30)
for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(cosine_sim_matrix)
    wcss.append(kmeans.inertia_)  

# Plot the Elbow Method graph
plt.figure(figsize=(10, 6))
plt.plot(K, wcss, 'bx-')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Within-Cluster Sum of Squares (WCSS)')
plt.title('Elbow Method For Optimal k')
plt.xticks(np.arange(min(K), max(K)+1, 1))
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import silhouette_score

silhouette_scores = []
cluster_range = range(2, 30)  # You can adjust the range as needed

for n_clusters in cluster_range:
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    cluster_labels = kmeans.fit_predict(cosine_sim_matrix)
    silhouette_avg = silhouette_score(cosine_sim_matrix, cluster_labels)
    silhouette_scores.append(silhouette_avg)
    
# Plot silhouette scores
plt.figure(figsize=(10, 6))
plt.plot(cluster_range, silhouette_scores, marker='o')
plt.title('Silhouette Score vs Number of Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.grid(True)
plt.show()
```

```python
from sklearn.cluster import KMeans

# Apply k-means clustering
kmeans = KMeans(n_clusters=15, random_state=42)
clusters = kmeans.fit_predict(cosine_sim_matrix)

# Add cluster labels to the DataFrame
risk_returns_all_yrs['Cluster'] = clusters

# Plotting the clusters
plt.figure(figsize=(10, 6))
plt.scatter(risk_returns_all_yrs['risk_level'], risk_returns_all_yrs['returns_1yr'], c=risk_returns_all_yrs['Cluster'], cmap='rainbow', alpha=0.7)
plt.title('K-means Clustering')
plt.xlabel('Risk Level')
plt.ylabel('Returns')
plt.grid(True)
plt.show()

print(risk_returns_all_yrs)
```
